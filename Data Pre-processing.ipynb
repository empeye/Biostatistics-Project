# Data Pre-processing

## Import .csv file

```python
import pandas as pd
data  = pd.read_csv('3_HCC_FibroScan_Data.csv')
data
```

## Information of data

```python
data.info()
```

```python
data.isnull().sum()
```

## 1. Drop null value

```python
data = data.dropna()
data
```

## 2. Change columns name

```python
data = data.rename(columns = {"lab_visit_vs13": "Lab_Visit", 
                                  "lab1_vs13":"Date", 
                                  "lab1_2_vs13":"SGPT",
                                  "lab2_2_vs13":"AFP",
                                  "lab3_2_vs13":"HBV_Viral_Load",
                                  "lab4_2_vs13":"HBV_log_equivalence",
                                  "lab5_3_vs13":"HBeAg",
                                  "class_lab5_3_vs13":"class_HBeAg",
                                  "fibo1_vs13": "Stiffness"})
data
```

## 3. Change Data Type from object to Date

```python
data['Date']= pd.to_datetime(data['Date'])
data.info()
```

## 4. Replace HBV Viral Load Data (<20 or >170M) and change data type to int

```python
data['HBV_Viral_Load'] = data['HBV_Viral_Load']\
    .str.replace("< 20","0")
data['HBV_Viral_Load'] = data['HBV_Viral_Load']\
    .str.replace("> 170000000","170000000")
data['HBV_Viral_Load']= data['HBV_Viral_Load']\
    .astype(int)
data.info()
```

## 5. Replace space from data in 'HBeAg' column

```python
data['HBeAg'] = data['HBeAg'].str.replace(" ","")
data
```

## Add column to classify group of data

```python
data['SGPT_Cat']=data['SGPT'].between(0, 40.1, inclusive=False).replace({0:'Abnormal',1:'Normal'})
```

```python
data['AFP_Cat']=data['AFP'].between(0, 7.1, inclusive=False).replace({0:'Abnormal',1:'Normal'})
```

```python
import numpy as np
bins=[0,8.1,10.1,14.1,np.inf]
names=['No to Mild','Moderate','Severe','Advanced']
data['Stiffness_Cat'] = pd.cut(data['Stiffness'], bins, labels=names)
data
```
